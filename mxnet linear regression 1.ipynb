```python
%matplotlib inline
import d2l
from mxnet import autograd, np, npx, nd
import mxnet as mx
import random
npx.set_np()
```

```python
l = np.array(lab)
labels = l.astype('float32') 
print(labels)
#print(l.dtype)
#print(type(l))
```

```python
f = np.array(feat)
features = f.astype('float32') 
print(features)
#print(f.dtype)
#print(type(f))
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # The examples are read at random, in no particular order
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = np.array(
            indices[i: min(i + batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]
```

```python
batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

```python
w = np.random.normal(0, 1e-10, (32,1))
b = np.zeros(1)
```

```python
w.attach_grad()
b.attach_grad()
```

```python
# Saved in the d2l package for later use
def linreg(X, w, b):
    regr = np.dot(X, w) + b
    #print('X', X)
    #print('Regr', regr)
    return regr
```

```python
print(w.shape, b.shape)
print(features.shape, labels.shape)
# f* w + b = l
```

```python
# Saved in the d2l package for later use
def squared_loss(y_hat, y):
    #print('regr output', y_hat)
    #print('yhat - y:',(y_hat - y.reshape(y_hat.shape))** 2 / 2  )
    sq_loss = (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
    return sq_loss
```

```python
# Saved in the d2l package for later use
def sgd(params, lr, batch_size):
    for param in params:
        param[:] = param - lr * param.grad / batch_size
```

```python
# In our case, big learning rate gets the gradient large very fast, so the output of the linear regression becomes infinite
# therefore the loss function becomes infinite as well
# until 1e-7 everything is ok, but if you go down to 1e-6, the gradient starts exploding again!
lr = 1e-7  # Learning rate

# Smaller learning rate means we learn slowly, so we can increase the number of epochs, until the point 
# the loss starts increasing again
num_epochs = 100  # Number of iterations

net = linreg  # Our fancy linear model
loss = squared_loss  # 0.5 (y-y')^2
```

```python
for epoch in range(num_epochs):
    # Assuming the number of examples can be divided by the batch size, all
    # the examples in the training dataset are used once in one epoch
    # iteration. The features and tags of minibatch examples are given by X
    # and y respectively
    for X, y in data_iter(batch_size, features, labels):
        with autograd.record():
            l = loss(net(X, w, b), y)  # Minibatch loss in X and y
        l.backward()  # Compute gradient on l with respect to [w, b]
        sgd([w, b], lr, batch_size)  # Update parameters using their gradient
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f' % (epoch + 1, train_l.mean().asnumpy()))
```
