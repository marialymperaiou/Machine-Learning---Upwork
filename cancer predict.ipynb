```python
import urllib.request
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read data from source
encoding = 'utf-8'
with urllib.request.urlopen("http://mlr.cs.umass.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wpbc.data") as url:
    s = url.read()
    # Convert data from byte str to str
    s = s.decode(encoding)
```

```python
# Python code to convert string to list 
def Convert(s, ch): 
    l = list(s.split(ch)) 
    return l 
```

```python
# Convert str to list. Each line of the str becomes an element of the list
text = Convert(s,'\n')
```

```python
# convert list to a list of lists
l = [] 
ch = ', '      # comma AND space!
for i in text: 
    sub = Convert(i,ch)
    l.append(sub)
```

```python
ch = ","                      # this char separates each attribute from another
new_l = []                    # new list that contains the attributes in different strings
for st in l:
    # st[0] refers to the str in each list of lists
    new_l.append(Convert(st[0],ch)) # split the str in many str, using the cmma as the separator
```

```python
df = pd.DataFrame(new_l)
```

```python
# Drop missing rows
df.dropna()
```

```python
# Find the indices of the rows where the outcome is N (= nonrecur)
# The second column (df[1]) indicates that
indexNames = df[df[1] == 'N'].index
print(indexNames)
```

```python
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
```

```python
# It is easier to have adjacent indices
df = df.reset_index()
print(df)
```

```python
df[:] = df[:].astype(str)
```

```python
# Keep only features for now
features = df.iloc[ : , 3:36 ]
print(features)
```

```python
idx = []     # this list will contain the row indices that contain non decimal elements
for i in range(len(features)):
    # eliminate decimal points, so that we can check if they are digits
    features.iloc[i,:]= features.iloc[i,:].str.replace(".", "") 
    # check if all columns are decimal
    if not(features.iloc[i].str.isdigit().all()):
        idx.append(i)
```

```python
# Delete these row indexes from dataFrame
df.drop(df.index[idx], inplace=True)
print(df)
```

```python
for i in range(len(df)):
    # Convert features from str to float
    df.iloc[i,3:36]= df.iloc[i,3:36].astype(float) 
```

```python
arr = features.values
arr = arr.astype('float64') 
print(arr)
```

```python
data=pd.DataFrame(arr) 

corr = data.corr()
fig, ax = plt.subplots(figsize=(20,20)) 

ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(10, 220, sep=80, n=7),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
   
    horizontalalignment='right'
);
```

```python
days = data.iloc[ : , 0]  # this is going to be our label that we are going to predict
```

```python
print(days)
```

```python
feat = data.iloc[ : , 1:8]  # this is going to be our label that we are going to predict
```

```python
#print(feat)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
X_train, X_test, Y_train, Y_test = train_test_split(feat, days, test_size = 0.50, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
Y_pred = regr.predict(X_test)
print('Coefficients: \n', regr.coef_)
print("Mean squared error: %.2f" % mean_squared_error(Y_test, Y_pred))
print("The labels: \n", np.array(Y_test).astype(int))
print("The predicted values are: \n", Y_pred.astype(int))
```
